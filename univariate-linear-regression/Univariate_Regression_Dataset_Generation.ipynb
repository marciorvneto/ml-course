```python
import numpy as np

N = 1000
x = np.linspace(100, 800, N)
e = np.random.randn(N) * 50
y = 0.3 * x + 120 + e 
```

```python
import matplotlib.pyplot as plt

plt.plot(x,y,'o')
```

```python
import pandas as pd

# Shuffle the indices and export the dataset

indices = np.arange(N)
np.random.shuffle(indices)

dataset = pd.DataFrame(np.column_stack((x[indices], y[indices])))

dataset.to_csv("apartment_data.csv", sep=',', index=False)
```
