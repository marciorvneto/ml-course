```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

font = {'family' : 'normal',
        'weight' : 'normal',
        'size'   : 16}

matplotlib.rc('font', **font)

def logistic(z):
    return 1/(1 + np.exp(-z))

z = np.linspace(-8,8,100)
y = logistic(z)

fig = plt.figure()

plt.plot(z,y)
plt.grid()
plt.yticks((0,0.5,1))
plt.xlabel(r"$z$")
plt.ylabel(r"$\sigma(z)$")
plt.title("The logistic function")

fig.savefig("logistic.png")
```
