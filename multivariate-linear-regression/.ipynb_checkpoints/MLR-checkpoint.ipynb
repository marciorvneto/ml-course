```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(5)

N = 100

x1 = np.random.random(N)* 600 + 100
x2 = np.random.random(N)*5 + 0.3

real_function = lambda x1,x2: 2*x1 - 0.3*x1*x2 - x2

y = np.zeros(N)
for i in range(N):
    y[i] = real_function(x1[i], x2[i])

df = pd.DataFrame({
    "area": x1,
    "distance": x2,
    "price": y
})

df.

df
```

```python
pd.plotting.scatter_matrix(df)
plt.suptitle("Scatter plot matrices for the apartment dataset")
```

```python
# fig = plt.figure()

# ax = fig.add_subplot(projection='3d')
# ax.scatter3D(x1, x2, y, marker="o", c=y)

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')
ax.view_init(10, 55)

ax.zaxis.set_rotate_label(False)  # disable automatic rotation
ax.set_zlabel('Z', rotation=90)  # rotate manually

ax.set_xlabel("Area")
ax.set_ylabel("Distance")
ax.set_zlabel("Price")

ax.scatter3D(x1, x2, y, marker="o", c=y, alpha=0.8)
```

```python
A = np.column_stack((np.ones(N), x1, x2))
coeffs = np.linalg.solve(A.T.dot(A), A.T.dot(y))
coeffs
```

```python
x1s = np.linspace(100,700,10)
x2s = np.linspace(0,6,10)
[X1, X2] = np.meshgrid(x1s,x2s)
Z = coeffs[0] + coeffs[1]*X1 + coeffs[2]*X2

print(Z.shape)

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')
ax.view_init(10, 55)

ax.scatter3D(x1, x2, y, marker="o", c="red", alpha=1)
ax.plot_surface(X1, X2, Z, alpha=0.5, color="blue")

ax.zaxis.set_rotate_label(False)  # disable automatic rotation
ax.set_zlabel('Z', rotation=90)  # rotate manually

ax.set_xlabel("Area")
ax.set_ylabel("Distance")
ax.set_zlabel("Price")
ax.set_title("Fitting a plane to a dataset", y=0.95)

for i in range(N):
    x1p = x1[i]
    x2p = x2[i]
       
    ypred = coeffs[0] + coeffs[1]*x1p + coeffs[2]*x2p
    ax.plot([x1p,x1p],[x2p,x2p],[ypred, y[i]], color="gray")
```
